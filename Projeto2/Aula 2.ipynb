```python
import yfinance as yf
acao = input('Digite o código da ação: ')
dados = yf.Ticker(acao).history("6mo")
fechamento = dados.Close
fechamento
```

```python
fechamento.plot()
```

# Passo 2

- Cotação Máxima
- Cotação Mínima
- Cotação Atual


```python
maxima = round(fechamento.max(), 2)
minima = round(fechamento.min(), 2)
atual = round(fechamento[-1], 2)
print(maxima)
print(minima)
print(atual)
```

# Passo 3
- Abrir uma nova aba (ctrl + t)
- Digitar www.gmail.com e dar enter
- Clicar no botão escrever
- Adicionar um destinatário (tab)
- Digitar o assunto do Email (tab)
- Preencher o corpo do Email
- Clicar no botão Enviar

```python
!pip install pyperclip
```

```python
import pyautogui as pg
import pyperclip as ppclip
import time
```

```python
# adicionar um timer para desacelerar o processo
pg.PAUSE = 4

# abri uma nova aba
pg.hotkey("ctrl", "t")

# digitar  www.gmail.com e dar enter

ppclip.copy("www.gmail.com")
pg.hotkey("ctrl", "v")
pg.hotkey("enter")

# clicar no botão Escrever
                                    # obs: time.sleep(5)
                                    # obs: pg.position() pega as coordenadas ao ficar com o mouse em cima
pg.click(x=50, y=198)

# adicionar o destinatário

ppclip.copy("jbarretofalcao@gmail.com")
pg.hotkey("ctrl", "v")
pg.hotkey("tab")


# digitar o assunto

ppclip.copy("Análise das Cotações da ação do dia")
pg.hotkey("ctrl", "v")
pg.hotkey("tab")

# reencher o corpo do Email
mensagem = """Prezado João, Ignore esse Email. Ele é um mero teste automático. Obrigado Pela Compreensão

    Segue, conforme solicitado, as análises da estudada nos últimos 6 meses
    
    Ação: {}

    Cotação Máxima: {}
    Cotação Mínima: {}
    Cotação Atual: {}
    
    Qualquer dúvida, fico à disposição para saná-la!!
    
    Atenciosamente, Perilo Oliveira
    
""".format(acao, maxima, minima, atual)

ppclip.copy(mensagem)
pg.hotkey("ctrl", "v")

# Clicar no botão Enviar
pg.click(x=848, y=694)

print("Email Enviado com Sucesso!!!!")
```
